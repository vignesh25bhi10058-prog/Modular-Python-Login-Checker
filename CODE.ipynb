```python
%%writefile credentials.py

# credentials.py
_VALID_USERS = {
    "admin": "Secure!123",
    "user1": "Password@456",
    "guest": "welcome1",
}
def get_all_users():
    return _VALID_USERS
```

```python
%%writefile validation.py
#validations
import re
from credentials import get_all_users
def is_username_valid(username):
    if len(username) < 4:
        return False, "Username must be at least 4 characters long."
    return True, "Username valid."

def is_password_secure(password):
    if len(password) < 8:
        return False, "Password must be at least 8 characters."
    if not re.search(r"[A-Z]", password):
        return False, "Password must contain an uppercase letter."
    if not re.search(r"[a-z]", password):
        return False, "Password must contain a lowercase letter."
    if not re.search(r"\d", password):
        return False, "Password must contain a digit."
    if not re.search(r"[!@#$%^&*()]", password):
        return False, "Password must contain a special character (!@#$%^&*())."
    return True, "Password secure."

def check_login_credentials(username, password):
    users = get_all_users()
    if username in users:
        if password == users[username]:
            return True, "Login Successful! Welcome back."
        else:
            return False, "Login Failed: Invalid password."
    else:
        return False, "Login Failed: Username not found."
```

```python
# main_app.py
from validation import (
    check_login_credentials,
    is_username_valid,
    is_password_secure
)
from credentials import get_all_users
def get_user_input(prompt):
    return input(f"{prompt}: ").strip()

def display_message(message):
    print(f"\n[INFO] {message}")

def run_login_system():
    display_message("--- User Login System ---")
    MAX_TRIES = 3
    attempt_count = 0

    while attempt_count < MAX_TRIES:
        username = get_user_input("Enter username")
        password = get_user_input("Enter password")

        is_successful, message = check_login_credentials(username, password)

        display_message(message)

        if is_successful:
            display_message("Access Allowed")
            return True
        else:
            attempt_count += 1
            remaining = MAX_TRIES - attempt_count
            if remaining > 0:
                display_message(f"Login failed. {remaining} attempt(s) remaining.\n")
            else:
                display_message("Access not allowed . all attempts have been used")
                return False
if __name__ == "__main__":
    run_login_system()
    display_message("System Check Complete")
```
